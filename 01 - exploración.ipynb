```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train = pd.read_csv("train.csv")
estrato_counts = train["F_ESTRATOVIVIENDA"].value_counts()
print("Cantidad de estudiantes por estrato:")
print(estrato_counts)
dept_counts = train["E_PRGM_DEPARTAMENTO"].value_counts().head(20)
plt.figure(figsize=(10,6))
sns.barplot(x=dept_counts.values, y=dept_counts.index, palette="viridis")
plt.title("Top 20 Departamentos con más estudiantes")
plt.xlabel("Número de estudiantes")
plt.ylabel("Departamento")
plt.show()
plt.figure(figsize=(8,5))
sns.barplot(x=estrato_counts.index, y=estrato_counts.values, palette="Set2")
plt.title("Cantidad de estudiantes por estrato de vivienda")
plt.xlabel("Estrato de vivienda")
plt.ylabel("Número de estudiantes")
plt.show()
```
